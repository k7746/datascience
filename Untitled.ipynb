```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv('examSet.csv')
```

```python
data
```

```python
data_list = data.values
data_list = data_list.tolist()
```

```python
for i in range(44):
    one_data = data_list[i]
    gas_month = one_data[31:61]
    cnd_month = one_data[61:91]
    hrs_month = one_data[91:121]
    plt.plot(gas_month, label = 'gas_month')
    plt.plot(cnd_month, label = 'cnd_month')
    plt.plot(hrs_month, label = 'hrs_month')
    plt.grid()
    plt.title(one_data[0])
    plt.xlabel("month")
    plt.ylabel("production")
    plt.show()
```
