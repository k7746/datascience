```python
def z_score_normalize(lst):
    normalized = []
    for value in lst:
        normalized_num = (value - np.mean(lst)) / np.std(lst)
        normalized.append(normalized_num)
    return normalized
```

```python
ans = [[0] * 2] * 29
num = 0
min = 2e9
for i in range(29):
    for j in range(232): # num of train
        for k in range(112): # num of 요인
            num = num + abs(new_data_n[k][i+232] - new_data_n[k][j])
        if min > num and num > 0:
            min = num
            row = j
        num = 0
    ans[i][0] = min
    ans[i][1] = row
    min = 2e9
    num = 0
print(ans)
```
